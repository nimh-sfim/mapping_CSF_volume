```python
import pandas as pd
import hvplot.pandas
import os.path as osp
```

```python
PRJ_DIR = '/data/SFIM_CSF_Volume/Project01'
PRCSDATA_DIR = osp.join(PRJ_DIR, 'PrcsData')

SBJ = 'SBJ05'
SES = 'SES01'
RUN = 'run3'
```

```python
THIS_DATA_PRCS_DIR = osp.join(PRCSDATA_DIR, SBJ, SES, RUN)
mot_estimates_refTI2TE1 = pd.read_csv(osp.join(THIS_DATA_PRCS_DIR, 'dfile.motion.1D'),header=None, sep='\\s+')
mot_estimates_refTI2TE1.columns = ['roll', 'pitch', 'yaw', 'dS', 'dL', 'dP']
mot_estimates_refTI2TE1.hvplot(title='Motion using TI2TE1 as reference')
                                      
```

```python
mot_estimates_selfref = {}
for suffix in ['TI1_TE1','TI1_TE2','TI1_TE3','TI2_TE1','TI2_TE2','TI2_TE3','TI3_TE1','TI3_TE2','TI3_TE3']:
    mot_estimates_selfref[suffix] = pd.read_csv(osp.join(THIS_DATA_PRCS_DIR, f'dfile.motion.{suffix}.1D'),header=None, sep='\\s+')
    mot_estimates_selfref[suffix].columns = ['roll', 'pitch', 'yaw', 'dS', 'dL', 'dP']
```

```python
(mot_estimates_selfref['TI2_TE1'] - mot_estimates_refTI2TE1).hvplot(title='Difference: Self-reference TI2TE1 minus reference TI2TE1')
```

```python
(mot_estimates_selfref['TI1_TE3'] - mot_estimates_refTI2TE1).hvplot(title='Difference in Motion Estimates: Self-reference TI1TE3 minus reference TI2TE1')
```
